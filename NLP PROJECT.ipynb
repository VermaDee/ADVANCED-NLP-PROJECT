```python
#importing all related libraries
import pandas as pd
import re 
import string
import syllables
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
nltk.download('stopwords')
```

```python
#text cleaner using regex
def text_cleaner(text):
    if not isinstance(text, float):
        text = text.lower()
        text = text.replace('\n', ' ')
        text = re.sub('\[.*?\]','',text) 
        text = re.sub('[%s]' % re.escape(string.punctuation),'',text)
        text = re.sub('\w*\d\w*','',text)
        text =re.sub(r'[“”–’]', '',text)
        text =re.sub(r'\b\w+\s*\s*\w+\b','',text)
    else:
        pass
    return text

cleaner = lambda x: text_cleaner(x)
```

```python
encoding='latin-1'
delimiter_pattern = r'\!|\?|\.'

df =  pd.read_csv('C:/Users/deevy/Desktop/scrape.txt', header=None, encoding=encoding, sep=delimiter_pattern)
df1 = pd.read_csv('C:/Users/deevy/Desktop/scrape.txt', header=None, encoding=encoding, sep=delimiter_pattern)
df2 = pd.read_csv('C:/Users/deevy/Desktop/scrape.txt', header=None, encoding=encoding, sep=delimiter_pattern)

df = df.dropna(axis=1, how='all')
df1 = df1.dropna(axis=1, how='all')
df2 = df2.dropna(axis=1, how='all')
```

```python
df
```

```python
for i in df.columns:
    df[i] = pd.DataFrame(df[i].apply(cleaner))
```

```python
stop_words = set(stopwords.words('english'))
for m in df.columns:
    df[m] = df[m].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in stop_words]))
```

```python
df
```

```python
#tokenization
for j in df.columns:
    df[j] = df[j].apply(lambda x: word_tokenize(x))

for j in df1.columns:
    df1[j] = df1[j].apply(lambda x: nltk.word_tokenize(x))
```

```python
 #CALCULATUIONS

#words after removing stopwords
words_cleaned=0
for column in df.columns:
    for token in df[column]:
        for tkn in token:
            words_cleaned=words_cleaned+1

#words before removing stopwords
words=0
for column in df1.columns:
    for token in df1[column]:
        for tkn in token:
            words=words+1

#number of sentences using df2
sentences=0
for column in df2.columns:
    sentences=sentences+1

#calculating the total number of characters using tokens from df1
total_characters = 0
for column in df1.columns:
    for token in df[column]:
        for tkn in token:
            total_characters = total_characters + len(tkn)

#average word length
Average_Word_Length=total_characters/words

#counting the number of persnal Pronouns
#(?!US\b) to avoid using country US as personal pronoun
pattern = r'\b(?!US\b)(he|she|it|they|we|you|I|me|him|her|us|them)\b'
pers_pro = 0
for column in df1.columns:
    for token in df1[column]:
        for tkn in token:
            if re.match(pattern, tkn):
                pers_pro=pers_pro+1

#counting the number of syllables using syllable library
sy_count=0
for column in df1.columns:
    for token in df1[column]:
        for tkn in token:
            syllable_count = syllables.estimate(tkn)
            if tkn.endswith("es") or tkn.endswith("ed"):
                syllable_count -= 1
            sy_count=sy_count+syllable_count

#counting the number of complex words
complex_count=0
for column in df1.columns:
    for token in df1[column]:
        for tkn in token:
            syllable_count = syllables.estimate(tkn)
            if tkn.endswith("es") or tkn.endswith("ed"):
                syllable_count -= 1

            #if a word has more than 2 syllable it is considered a complex word
            if syllable_count>2:
                complex_count=complex_count+syllable_count
```

```python
#column name
column_name=['Cleaned Words', 'Total Words','Sentences','Total Characters','Average Word Length', 'Personal Pronouns', 'Syllable', 'Complex Words']
```

```python
dataoutput=[]
dataoutput.append([words_cleaned,words,sentences,total_characters,Average_Word_Length,pers_pro,sy_count,complex_count])
```

```python
dfmain = pd.DataFrame(dataoutput, columns=column_name)
```
